```python
import pandas as pd
import numpy as np
```

```python
# Mengganti path dengan direktori yang sesuai
path = '/content/drive/My Drive/Collab Notebook/'

kumpulan_datajalan = pd.read_csv(path + 'Dataset Asli.csv', encoding="ISO-8859-1")
```

LOAD DATA

```python
features_datajalan = kumpulan_datajalan[["Sub Kategori", "Rusak Ruas Kiri Jalan", "Rusak Ruas Kanan Jalan", "Nama Wilayah","Panjang (M)", "Lebar (M)", "Tebal/Tinggi (M)"]].copy()

# Menampilkan data dalam bentuk tabel
features_datajalan
```

```python
# Print contoh data kosong dalam bentuk tabel
data_cleaning = features_datajalan.dropna()
data_cleaning
```

```python
data_cleaning.isnull().sum()
```

```python
# Mengonversi kolom 'Local_Amount' ke tipe data string
data_cleaning['Panjang (M)'] = data_cleaning['Panjang (M)'].astype(str)
data_cleaning['Lebar (M)'] = data_cleaning['Lebar (M)'].astype(str)
data_cleaning['Tebal/Tinggi (M)'] = data_cleaning['Tebal/Tinggi (M)'].astype(str)

# Menghilangkan koma dari kolom 'Local_Amount'
data_cleaning['Panjang (M)'] = data_cleaning['Panjang (M)'].str.replace(',', '').astype(float)
data_cleaning['Lebar (M)'] = data_cleaning['Lebar (M)'].str.replace(',', '').astype(float)
data_cleaning['Tebal/Tinggi (M)'] = data_cleaning['Tebal/Tinggi (M)'].str.replace(',', '').astype(float)

# Menampilkan data hasil akhir yang sudah diubah menjadi float
data_cleaning[['Panjang (M)', 'Lebar (M)', 'Tebal/Tinggi (M)']]
```

```python
# Multiply 'jumlah_projek' with 'nilai_projek' and store the result in a new column 'total_nilai'
data_cleaning['Volume Kerusakan'] = data_cleaning['Panjang (M)'] * data_cleaning['Lebar (M)' ]* data_cleaning['Tebal/Tinggi (M)']

# Print the updated DataFrame
data_cleaning
```

```python
# Fungsi untuk menentukan
def sub_kategori_kerusakan(Kategori):
    if Kategori in ['111 (Lubang)', '112 (Keriting)', '114 (Amblas)','116 (Kerusakan Tepi)', '117 (Retak Buaya)', '118 (Retak Garis)', '119 (Kegemukan Aspal)','120 (Terkelupas)', '160 (Permukaan Aspal Kasar)','232 (Alur)','252 (Kehilangan Permukaan)','612 (Tergerus)','710 (Longsor)']:
        return 'Rusak'
    elif Kategori in ['115 (Jembul)', '253 (Rumput Panjang)', '412 (Kerusakan Saluran)','413 (Saluran Ditumbuhi Tanaman)','431 (Pendangkalan)','417 (Tersumbat)','511 (Patok KM Rusak)','531 (Marka Jalan Pudar)','640 (Pas. Batu Amblas)','641 (Kehilangan Batuan)','840 (Kerusakan Sayap Gorong-gorong)','Kondisi Baik']:
        return 'Tidak Rusak'
    else:
        return 'Tidak Diketahui'

# Terapkan fungsi pada kolom 'Jenis Kelamin'
data_cleaning['Kategori Kerusakan'] = data_cleaning['Sub Kategori'].apply(sub_kategori_kerusakan)

# Tampilkan DataFrame hasil
data_cleaning
```

```python
# Fungsi untuk menentukan kualitas berdasarkan rentang
def tentukan_volume(volume):
    if 0.000 <= volume <= 5620:
        return 'Volume Rusak'
    elif 5621 <= volume <= 675000:
        return 'Volume Tidak Rusak'
    else:
        return 'Volume Tidak Valid'

# Terapkan fungsi pada kolom 'Kualitas'
data_cleaning['Kategori Volume Kerusakan'] = data_cleaning['Volume Kerusakan'].apply(tentukan_volume)

# Tampilkan DataFrame hasil
data_cleaning
```
